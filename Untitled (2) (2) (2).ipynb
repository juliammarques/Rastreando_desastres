<h3>Preparação dos dados</h3>
Considerando que a ideia principal dessa analise é determinar o que pode ou não pode ser um desastre, para que as autoridades possam interferir, não faz sentido analisar os tweets sem localização, portanto nessa etapa iremos excluir os tweets com a localização em branco e limpar os dados.

```python
twitter_train.dropna(inplace=True)
```

<h4>Limpando o texto</h4>

```python
# remover links e url
def exclui_URL(text):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'',text)
# remover pontuação
def exclui_pontuacao(text):
    table=str.maketrans('','',string.punctuation)
    return text.translate(table)
# remover %20
def limpar_keyword(text):
    text = text.replace('%20',' ') 
    return text
```

```python
twitter_train['text']=twitter_train['text'].apply(lambda x : exclui_URL(x))
twitter_train.head(10)
```

```python
twitter_train['text']=twitter_train['text'].apply(lambda x : exclui_pontuacao(x))
twitter_train.head(10)
```

```python
twitter_train['keyword']=twitter_train['keyword'].apply(lambda x : limpar_keyword(x))
twitter_train['keyword'].unique()
```

```python
twitter_train['keyword'].nunique()
```

<h3>Transformação dos dados:</h3>
Nessa etapa para melhorar a compreensão do que estamos analisando, vamos alterar o nome da coluna target para desastre e o resultado 0(desastre ficticio) e 1(desastre real) para ficticio e real, de modo a facilitar a compreensão de quem esta lendo este documento.

```python
new_target=pd.Categorical(twitter_train["target"])
new_target=new_target.rename_categories(["Ficticio","Real"]) 
twitter_train["target"]= new_target
twitter_train.rename(columns={'target':'Desastre'}, inplace = True)
twitter_train.head(10)
```

<strong>Modelo de predição</strong>

```python
def testIfTrueOrFake(text):
    highRealPossibility = ['outbreak', 'wreckage', 'oil20spill', 'collision', 'evacuated','electrocute', 'blazing', 'crush', 'traumatised', 'screaming','ablaze']
    if(text in highRealPossibility):
        return 'real'
    else:
        return 'ficticio'
  
```

```python
twitter_train['Desastre']=twitter_train['keyword'].apply(lambda x : testIfTrueOrFake(x))
twitter_train.head(100)
```
